```python
import mysql.connector

mydb = mysql.connector.connect(
  host="localhost",
  user="root",
  passwd="Phylius@1990"
)

mycursor = mydb.cursor()

mycursor.execute("CREATE DATABASE sprint")
```

```python
mycursor.execute("CREATE TABLE sprint3 (items VARCHAR(255), `Selling Price` Double, `Stock Price` Double, Quantity Int )")
```

```python
#Insert chips
mycursor = mydb.cursor()

sql = "INSERT INTO sprint3 (items, `Selling Price`, `Stock Price`, Quantity) VALUES (%s, %s, %s, %s)"
val = ("Cornitos", 11, 9.50, 65)
mycursor.execute(sql, val)

mydb.commit()

print(mycursor.rowcount, "record inserted.")
```
